# Splitting the Data

```python
x = diabetes.iloc[:,:-1]
```

```python
y = diabetes.iloc[:,-1]
```

# Data Pre-processing

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x.shape,x_train.shape,x_test.shape
```

# Training Data

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
classifier.fit(x_train,y_train)
```

# Model Prediction

```python
y_preds = classifier.predict(x_train)
y_preds
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

# Comparison
* Comparing the test values and the predicted values:
* weather the Model is Overfitting or Underfitting

```python
classifier_dif = pd.DataFrame({'Actual value': y_train, 'Predicted value': y_preds})
classifier_dif
```

```python
classifier_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred})
classifier_diff
```

# Checking the Residuals

```python
residuals = y_test - y_pred
residuals
```

```python
residual = y_train - y_preds
residual
```

# Evaluating the model

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_preds,y_train)
'Accuracy Score Trained Data:',accuracy
```

```python
scores = accuracy_score(y_pred,y_test)
'Accuracy Score Tested Data:',scores
```

**Loading the Machine Learning Model**

```python
# load the save model
load_model = pickle.load(open('/content/trained _model.py','rb'))

# create a function
def diabetes_prediction(input_data):

    # Changing the input_data_as_numpy_array
    input_data_as_numpy_array = np.asarray(input_data)

    # Now, we have to reshape the array because we are predicting for just single patient
    input_data_reshaped =input_data_as_numpy_array.reshape(1,-1)

    # Making Prediction
    prediction = load_model.predict(input_data_reshaped )

    if (prediction[0]) == 0:
      return'Not a Diabitic Person'
    else:
      return'Diabitic Person'
```
